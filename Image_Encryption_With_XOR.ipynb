```python
# prompt: create a function to upload 2 image a normal image and a key image and use the funtion above to do exclusive xor with it.

from google.colab import files
from io import BytesIO
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def input_image():
    uploaded = files.upload()
    for fn in uploaded.keys():
        image = Image.open(BytesIO(uploaded[fn]))
        return image
    return None  # Return None if no file is uploaded

def convert_to_grayscale(image):
    if image:
        return image.convert("L")
    return None

def resize_image_640x640(image):
    width, height = image.size
    if width > height:
        new_width, new_height = 640, int(height * (640 / width))
    else:
        new_width, new_height = int(width * (640 / height)), 640
    resized_image = image.resize((new_width, new_height))

    new_image = Image.new("L", (640, 640), 255)
    paste_x, paste_y = (640 - new_width) // 2, (640 - new_height) // 2
    new_image.paste(resized_image, (paste_x, paste_y))
    return new_image

def convert_grayscale_to_pixel(image):
    if image:
        return list(image.getdata())
    return None

def convert_pixels_to_2d_array(pixels, width, height):
    if pixels:
        return [pixels[y * width:(y + 1) * width] for y in range(height)]
    return None

def convert_2d_to_1d_array(image_2d):
    if image_2d:
        return np.array(image_2d).flatten()
    return None

def convert_flatten_numpy_to_binary(flatten_numpy_array):
    if flatten_numpy_array is not None:
        threshold = 128  # You can adjust this threshold
        return [1 if pixel >= threshold else 0 for pixel in flatten_numpy_array]
    return None

def process_images():
    # Load the first (normal) image
    print("Upload the first (normal) image:")
    normal_image = input_image()

    # Load the second (key) image
    print("Upload the second (key) image:")
    key_image = input_image()

    # Convert both images to grayscale
    grayscale_normal = convert_to_grayscale(normal_image)
    grayscale_key = convert_to_grayscale(key_image)

    # Resize both images to 640x640
    resized_normal = resize_image_640x640(grayscale_normal)
    resized_key = resize_image_640x640(grayscale_key)

    # Convert to pixel lists
    normal_pixels = convert_grayscale_to_pixel(resized_normal)
    key_pixels = convert_grayscale_to_pixel(resized_key)

    # Convert normal pixels to 2D array
    normal_2d = convert_pixels_to_2d_array(normal_pixels, 640, 640)
    key_2d = convert_pixels_to_2d_array(key_pixels, 640, 640)

    # Convert 2D arrays to 1D
    normal_1d = convert_2d_to_1d_array(normal_2d)
    key_1d = convert_2d_to_1d_array(key_2d)

    # Convert to binary arrays
    normal_binary = convert_flatten_numpy_to_binary(normal_1d)
    key_binary = convert_flatten_numpy_to_binary(key_1d)

    # Perform XOR on binary values
    xor_result = [normal_bit ^ key_bit for normal_bit, key_bit in zip(normal_binary, key_binary)]

    # Convert XOR result back to grayscale image
    xor_result_image = Image.new("L", (640, 640))
    xor_result_image.putdata([bit * 255 for bit in xor_result])  # Scale to 0-255

    # Display the encrypted XOR result
    print("Encrypted image:")
    plt.imshow(xor_result_image, cmap='gray')
    plt.axis('off')
    plt.show()

    # Ask the user if they want to decrypt
    decrypt_choice = input("Do you want to decrypt the image back to the original? (yes/no): ")
    if decrypt_choice.lower() == 'yes':
        # Decrypt by XORing the encrypted binary image with the key binary
        decrypted_binary = [encrypted_bit ^ key_bit for encrypted_bit, key_bit in zip(xor_result, key_binary)]

        # Convert decrypted binary back to image
        decrypted_image = Image.new("L", (640, 640))
        decrypted_image.putdata([bit * 255 for bit in decrypted_binary])  # Scale to 0-255

        # Display the decrypted image
        print("Decrypted image (should match the original):")
        plt.imshow(decrypted_image, cmap='gray')
        plt.axis('off')
        plt.show()

# Run the function
process_images()
```
